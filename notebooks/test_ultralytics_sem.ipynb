```python
from ultralytics import YOLO
from ultralytics.utils.plotting import colors  

# Custom palette (BGR format for OpenCV/Ultralytics)
custom_palette_bgr = [
    (0, 0, 0),       # Class 0: background -> Black
    (0, 255, 0),     # Class 1: cuttable -> Green (BGR: B=0, G=255, R=0)
    (255, 0, 0),     # Class 2: foliage -> Blue (BGR: B=255, G=0, R=0)
    (0, 0, 255),     # Class 3: non-traversable -> Red (BGR: B=0, G=0, R=255)
    (0, 255, 255),   # Class 4: traversable -> Yellow (BGR: B=0, G=255, R=255)
]


# Monkey-patch palette
colors.palette = custom_palette_bgr
colors.n = len(custom_palette_bgr)

# Load model
model = YOLO("yolo26n-sem.yaml").load("yolo26n-sem.pt")

# Train
results = model.train(
    data="../data/japan-pear-orchard-semantic-bg0/dataset.yaml",
    epochs=100,
    imgsz=1024,
    name="pear_orchard_sem"
)
```
